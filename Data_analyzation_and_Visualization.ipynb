# **Groupby**

Groupby method allows us to group rows of data together & call aggregate function.

```python
data = {'company':['A','A','B','B','C','C','C'],
        'person':['sam','charlie','amy','vanessa','carl','sarah','allen'],
        'sales':[180,340,260,170,150,290,330]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('company').sum()
```

```python
df.groupby('company').min()
```

```python
df.groupby('company').max()
```

```python
df.groupby('company').mean()
```

```python
df.groupby('company').count()
```

# **Merging, Joining, Concatenating**

```python
df1 = pd.DataFrame({'A': ['A0','A1','A2','A3'],
                    'B': ['B0','B1','B2','B3'],
                    'C': ['C0','C1','C2','C3'],
                    'D': ['D0','D1','D2','D3']},
                   index=[0,1,2,3])
```

```python
df2 = pd.DataFrame({'A': ['A4','A5','A6','A7'],
                    'B': ['B4','B5','B6','B7'],
                    'C': ['C4','C5','C6','C7'],
                    'D': ['D4','D5','D6','D7']},
                   index=[4,5,6,7])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

# **Merging**

```python
jan_data = pd.DataFrame({'cust_id':['C0','C1','C2','C3','C4'],
                         'jan_month_bill':[80,25,40,360,140]})
```

```python
feb_data = pd.DataFrame({'cust_id':['C0','C1','C2','C3','C5'],
                         'feb_month_bill':[280,251,240,30,340]})
```

```python
jan_data
```

```python
feb_data
```

```python
jan = pd.DataFrame({'customer_id':['C0','C1','C2','C3','C4'],
                         'jan_month_bill':[80,25,40,360,140]})

feb = pd.DataFrame({'cust_id':['C0','C1','C2','C3','C5'],
                         'feb_month_bill':[280,251,240,30,340]})
```

```python
jan.merge(feb,left_on='customer_id',right_on='cust_id')
```

# **Joining**

```python
df_maths_marks = pd.DataFrame({'maths_marks':[90,78,98,95,94],'maths_grade':[1,2,1,1,1]},index=['s1','s2','s3','s4','s6'])
```

```python
df_chemistry_marks = pd.DataFrame({'chemistry_marks':[88,58,99,91,95],'chemistry_grade':[2,3,1,1,1]},index=['s1','s2','s3','s4','s5'])
```

```python
df_maths_marks.join(df_chemistry_marks,how='outer')
```
